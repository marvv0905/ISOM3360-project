```python
from google.colab.widgets import TabBar
import ipywidgets as widgets
import time
from ipywidgets import interact
from ipywidgets import Text
from ipywidgets import DatePicker
from ipywidgets import IntText
from ipywidgets import Label
from datetime import date

tb_all = TabBar(['Introduction', 'Main Menu', 'Show All Students', 'Show 3 Students', 'Show Record Detail'])

# a simple function for jumping between tabs
def showPage(tabName):
  with tb_all.output_to(tabName):
    display()

##########################################################
################ Introduction Tab - Begin ################
##########################################################
with tb_all.output_to('Introduction'):

# create 1 label
  intro_label = Label("ISOM3260-Lab07p Sample Code")

# create 1 button
  btnContinue = widgets.Button(description="Continue")

# define the event handlers for the two buttons when they are being clicked
  def on_btnContinue_button_clicked(b):
    showPage("Main Menu")
  btnContinue.on_click(on_btnContinue_button_clicked)

# display the objects
  intro_row1HBox = widgets.HBox([intro_label])
  intro_row2HBox = widgets.HBox([btnContinue])

  display(widgets.VBox([intro_row1HBox, intro_row2HBox]))
################ Introduction Tab - End ################



#######################################################
################ Main Menu Tab - Begin ################
#######################################################
with tb_all.output_to('Main Menu'):

# create 1 label
  mainMenu_label = Label("Please select a function...")

# create 4 button
  btnShowAll   = widgets.Button(description="Show all students")
  btnShowThree = widgets.Button(description="Show 3 students")

# define the event handlers for the two buttons when they are being clicked
  def on_btnShowAll_button_clicked(b):
    showPage("Show All Students")
  btnShowAll.on_click(on_btnShowAll_button_clicked)

  def on_btnShowThree_button_clicked(b):
    showPage("Show 3 Students")
  btnShowThree.on_click(on_btnShowThree_button_clicked)

# display the objects
  mainMenu_row1HBox = widgets.HBox([mainMenu_label])
  mainMenu_row2HBox = widgets.HBox([btnShowAll])
  mainMenu_row3HBox = widgets.HBox([btnShowThree])

  display(widgets.VBox([mainMenu_row1HBox, mainMenu_row2HBox, mainMenu_row3HBox]))
################ Main Menu Tab - End ################



###############################################################
################ Show All Students Tab - Begin ################
###############################################################
with tb_all.output_to('Show All Students'):
# create 3 buttons and 1 label
  ret_all_button = widgets.Button(description="Retrieve All")
  ret_all_clear_button = widgets.Button(description="Clear All")
  ret_all_back_button = widgets.Button(description="Back")
  ret_all_spacer = widgets.Label("--------------------------------------------------------------------------")
  ret_all_out = widgets.Output()


# create 5 lists for containing Textboxes and 1 list for containing HBoxes
  ret_all_id_textboxes    = []
  ret_all_name_textboxes  = []
  ret_all_email_textboxes = []
  ret_all_year_textboxes  = []
  ret_all_dob_textboxes   = []
  ret_all_HBoxes = []


# define the event handlers for the buttons when they are being clicked
  def on_ret_all_button_clicked(b):
    on_ret_all_clear_button_clicked(b)

    c.execute("SELECT * FROM PYSTUDENT")
    for row in c:
      ret_all_id_textboxes.append(Text(value=str(row[0]), description='Student ID:', disabled=True))
      ret_all_name_textboxes.append(Text(value=str(row[1]), description='Name:', disabled=True))
      ret_all_email_textboxes.append(Text(value=str(row[2]), description='Email:', disabled=True))
      ret_all_year_textboxes.append(Text(value=str(row[3]), description='Year:', disabled=True))
      ret_all_dob_textboxes.append(Text(value=str(row[4]), description='DOB:', disabled=True))

    for i in range(len(ret_all_id_textboxes)):
      ret_all_HBoxes.append( widgets.HBox(  [ret_all_id_textboxes[i], ret_all_name_textboxes[i], ret_all_email_textboxes[i], ret_all_year_textboxes[i], ret_all_dob_textboxes[i]]  ) )

    with ret_all_out:
      display(widgets.VBox([i for i in ret_all_HBoxes]))
  ret_all_button.on_click(on_ret_all_button_clicked)


  def on_ret_all_clear_button_clicked(b):
    ret_all_id_textboxes.clear()
    ret_all_name_textboxes.clear()
    ret_all_email_textboxes.clear()
    ret_all_year_textboxes.clear()
    ret_all_dob_textboxes.clear()

    ret_all_HBoxes.clear()
    ret_all_out.clear_output()
  ret_all_clear_button.on_click(on_ret_all_clear_button_clicked)


  def on_ret_all_back_button_clicked(b):
    showPage("Main Menu")
  ret_all_back_button.on_click(on_ret_all_back_button_clicked)

# display the objects
  ret_all_control_HBox = widgets.HBox([ret_all_button, ret_all_clear_button])
  ret_all_VBox = widgets.VBox([ret_all_control_HBox, ret_all_spacer, ret_all_out, ret_all_back_button])
  display(ret_all_VBox)
################ Show All Students Tab - End ################



#############################################################
################ Show 3 Students Tab - Begin ################
#############################################################
with tb_all.output_to('Show 3 Students'):

# create 6 buttons and 2 label (1 hidden)
  ret_3_button = widgets.Button(description="Retrieve Now")
  ret_3_clear_button = widgets.Button(description="Clear")
  ret_3_prev_button = widgets.Button(description="Previous 3", disabled=True)
  ret_3_next_button = widgets.Button(description="Next 3", disabled=True)
  ret_3_back_button = widgets.Button(description="Back")
  ret_3_spacer = widgets.Label("--------------------------------------------------------------------------")
  current_record_index = widgets.Label("0")

# create 2 lists for containing Textboxes, 1 list for containing Detail Buttons, and 1 list for containing HBoxes
  ret_3_id_textboxes    = [Text(value='', description='Student ID:', disabled=True) for i in range(3)]
  ret_3_name_textboxes  = [Text(value='', description='Name:', disabled=True)       for i in range(3)]
  ret_3_detail_buttons  = [widgets.Button(description="More detail", disabled=True) for i in range(3)]
  ret_3_data_HBoxes = [ widgets.HBox(  [ret_3_id_textboxes[i], ret_3_name_textboxes[i], ret_3_detail_buttons[i]]  ) for i in range(3) ]

# define a function to retrieve students at defined indexes
  def showRecords(from_index=1, to_index=3):
    c.execute("SELECT COUNT(*) FROM PYSTUDENT")
    for row in c:
      max_record_index = row[0]

    if max_record_index == 0:
      ret_3_spacer.value = "No record found."
      return

    i = 0
    c.execute("SELECT rownum, student_id, student_name, email, year, dob FROM PYSTUDENT")
    for row in c:
      if from_index <= row[0] <= to_index:
        ret_3_id_textboxes[i].value = str(row[1])
        ret_3_name_textboxes[i].value = str(row[2])
        ret_3_detail_buttons[i].disabled = False

        i = i + 1
        if i > 2:
          break
      else:
        continue
    current_record_index.value = str(from_index)

    if from_index > 1: ret_3_prev_button.disabled = False
    if max_record_index > to_index: ret_3_next_button.disabled = False
    ret_3_spacer.value = "{} to {} --------------------------------------------------------------------------".format(from_index, to_index if max_record_index > to_index else max_record_index)


# define the event handlers for the buttons when they are being clicked
  def on_ret_3_button_clicked(b):
    from_index = 1
    to_index = from_index + 2

    on_ret_3_clear_button_clicked(b)
    showRecords(from_index, to_index)

  ret_3_button.on_click(on_ret_3_button_clicked)


  def on_ret_3_prev_button_clicked(b):
    from_index = int(current_record_index.value) - 3
    to_index = from_index + 2

    on_ret_3_clear_button_clicked(b)
    showRecords(from_index, to_index)
  ret_3_prev_button.on_click(on_ret_3_prev_button_clicked)


  def on_ret_3_next_button_clicked(b):
    from_index = int(current_record_index.value) + 3
    to_index = from_index + 2

    on_ret_3_clear_button_clicked(b)
    showRecords(from_index, to_index)
  ret_3_next_button.on_click(on_ret_3_next_button_clicked)


  def on_ret_3_back_button_clicked(b):
    showPage("Main Menu")
  ret_3_back_button.on_click(on_ret_3_back_button_clicked)


  def on_ret_3_clear_button_clicked(b):
    current_record_index.value = "0"
    max_record_index = 0
    ret_3_prev_button.disabled = True
    ret_3_next_button.disabled = True

    for i in range(3):
      ret_3_id_textboxes[i].value= ''
      ret_3_name_textboxes[i].value= ''
      ret_3_detail_buttons[i].disabled = True
    ret_3_spacer.value = "--------------------------------------------------------------------------"
  ret_3_clear_button.on_click(on_ret_3_clear_button_clicked)


  def on_ret_3_detail_button_clicked(b):
    # check which button is being clicked, you only have 3 buttons: 0, 1, 2
    button_num = 0
    for button_num in range(len(ret_3_detail_buttons)):
      if ret_3_detail_buttons[button_num] == b:   # button found!
        student_id_selected = ret_3_id_textboxes[button_num].value
        break

    # prepare the values in the detail page, and then move to the detail page
    dtl_id_textbox.value = str(student_id_selected)
    c.execute("SELECT * FROM PYSTUDENT WHERE student_id = '{}'".format(student_id_selected))
    for row in c:
      dtl_name_textbox.value = row[1]
      dtl_email_textbox.value = row[2]
      dtl_year_intbox.value = row[3]
      dtl_DOB_datepicker.value = row[4]
    showPage("Show Record Detail")

  # assign the event handler function to all More Detail buttons
  for dtl_btn in ret_3_detail_buttons:
    dtl_btn.on_click(on_ret_3_detail_button_clicked)


# display the objects
  ret_3_control_HBox  = widgets.HBox([ret_3_button, ret_3_clear_button])
  ret_3_rcd_ctrl_HBox = widgets.HBox([ret_3_prev_button, ret_3_next_button])
  ret_3_VBox_items = [ret_3_control_HBox, ret_3_spacer] + ret_3_data_HBoxes + [ret_3_rcd_ctrl_HBox, ret_3_back_button]

  ret_3_VBox = widgets.VBox(ret_3_VBox_items)
  display(ret_3_VBox)
################ Show 3 Students Tab - End ################



################################################################
################ Show Record Detail tab - Begin ################
################################################################
with tb_all.output_to('Show Record Detail'):

# create 6 label and 5 textboxes
  dtl_id_label = Label("Student ID: ")
  dtl_id_textbox = Text(value='', placeholder='Student ID', disabled=False)

  dtl_spacer = Label("---------------------------------------------------------------------------------------------------------------------------------")

  dtl_name_label = Label("Name: ")
  dtl_name_textbox = Text(placeholder='Name:', description='Name:', disabled=False)

  dtl_email_label = Label("Email: ")
  dtl_email_textbox = Text(description='Email:', disabled=False)

  dtl_year_label = Label("Year: ")
  dtl_year_intbox = IntText(description='Year:', disabled=False)

  dtl_dob_label = Label("DOB")
  dtl_DOB_datepicker = DatePicker(description='DOB:', disabled=False)

# create 4 buttons
  dtl_back_button = widgets.Button(description="Back")

# define the event handlers for the buttons when they are being clicked
  def on_dtl_back_button_clicked(b):
    showPage("Show 3 Students")
  dtl_back_button.on_click(on_dtl_back_button_clicked)

# display the objects
  dtl_line1HBox = widgets.HBox([dtl_id_label, dtl_id_textbox])
  dtl_line2HBox = widgets.HBox([dtl_spacer])
  dtl_textboxVBox = widgets.VBox([dtl_name_textbox, dtl_email_textbox, dtl_year_intbox, dtl_DOB_datepicker])
  dtl_line3HBox = widgets.HBox([dtl_textboxVBox])
  dtl_line4HBox = widgets.HBox([dtl_back_button])

  display(widgets.VBox([dtl_line1HBox, dtl_line2HBox, dtl_line3HBox, dtl_line4HBox]))
################ Delete Data Tab - End ################

showPage("Introduction")
```

```python
from google.colab.widgets import TabBar
import ipywidgets as widgets
import time
import pandas as pd
from ipywidgets import interact, Text, DatePicker, IntText, Label, Layout, Button, Box, GridspecLayout, Checkbox, Output
from datetime import date

tb_all = TabBar(['Introduction', 'Sandbox', 'MessageBox'])

# a simple function for jumping between tabs
def showPage(tabName):
  with tb_all.output_to(tabName):
    display()


##########################################################
################ Introduction Tab - Begin ################
##########################################################
with tb_all.output_to('Introduction'):

  intro_label = Label("ISOM3260-Lab06p Sample Code")
  btnContinue = widgets.Button(description="Continue")

  # define the event handlers for the buttons when they are being clicked
  def on_btnContinue_button_clicked(b):
    showPage("Sandbox")
  btnContinue.on_click(on_btnContinue_button_clicked)

  # display the objects
  intro_row1HBox = widgets.HBox([intro_label])
  intro_row2HBox = widgets.HBox([btnContinue])

  display(widgets.VBox([intro_row1HBox, intro_row2HBox]))

################ Introduction Tab - End ################



#####################################################
################ Sandbox Tab - Begin ################
#####################################################
with tb_all.output_to('Sandbox'):

  grid2 = GridspecLayout(9, 2, height='430px', width='650px', grid_gap='10px')

  ## Greeting Message Box
  btnMsgBox = widgets.Button(description="Greeting")

  # define the event handlers for the buttons when they are being clicked
  def on_btnMsgBox_button_clicked(b):
    currentDate = currentTime = greetMsg = ""

    c.execute("SELECT TO_CHAR(SYSDATE, 'dd/mm/yyyy') , TO_CHAR(SYSDATE, 'hh24:mi:ss') FROM DUAL")
    for row in c:
      currentDate = row[0]
      currentTime = row[1]

    if currentTime < "12":
      greetMsg = "Good morning! ^o^"
    elif currentTime > "12" and currentTime < "18":
      greetMsg = "Good afternoon! ^o^"
    else:
      greetMsg = "Good evening! ^o^"

    messageBox_label.value = "Today is " + currentDate + ". " + "It is now " + currentTime + ". " + greetMsg
    showPage("MessageBox")
  btnMsgBox.on_click(on_btnMsgBox_button_clicked)

  # display the objects
  hbRow1_layout = widgets.Layout(display='center', flex_flow='col', justify_content='center', align_items='center', border='1px solid grey', height='35px')
  hbRow1 = widgets.HBox([btnMsgBox], layout=hbRow1_layout)
  grid2[0,:] = hbRow1


  ## Checkbox / Vote
  voteTitle = Label("Which flavour(s) do you like?")
  cb1 = Checkbox(description='Cupcake', value=False, disabled=False, indent=False)
  cb2 = Checkbox(description='Donut', value=False, disabled=False, indent=False)
  cb3 = Checkbox(description='Ice-cream Sandwich', value=False, disabled=False, indent=False)
  cb4 = Checkbox(description='Marshmallow', value=False, disabled=False, indent=False)
  btnVote = widgets.Button(description="Vote")

  # define the event handlers for the buttons when they are being clicked
  def on_btnVote_button_clicked(b):
    msgString = ""
    if cb1.value == True:
      msgString = msgString + "Cupcake" + ", "
    if cb2.value == True:
      msgString = msgString + "Donut" + ", "
    if cb3.value == True:
      msgString = msgString + "Ice-cream Sandwich" + ", "
    if cb4.value == True:
      msgString = msgString + "Marshmallow" + ", "
    if len(msgString) > 0:
      messageBox_label.value = "I vote for " + msgString[0:(len(msgString)-2)] + "!!!"
      showPage("MessageBox")
  btnVote.on_click(on_btnVote_button_clicked)

  # display the objects
  vb1Col1_layout = widgets.Layout(display='flex-start', flex_flow='col', align_items='flex-start', border='1px solid grey')
  vb1Col1 = widgets.VBox([voteTitle, cb1, cb2, cb3, cb4, btnVote], layout=vb1Col1_layout)
  grid2[1:6,0] = vb1Col1


  ## Radio Button / Choose
  chooseTitle = Label("Which one do you like most?")
  rbGroup = widgets.RadioButtons(options=["Coke", "Cream Soda", "Fanta", "Say no to soft drink?"], value=None)
  btnChoose = widgets.Button(description="Choose")

  # define the event handlers for the buttons when they are being clicked
  def on_btnChoose_button_clicked(b):
    msgString = ""
    if rbGroup.value == "Coke":
      msgString = "Give me coke!"
    if rbGroup.value == "Cream Soda":
      msgString = "Can I have cream soda in swimming pool?"
    if rbGroup.value == "Fanta":
      msgString = "Fantastic fanta!"
    if rbGroup.value == "Say no to soft drink?":
      msgString = "Stay healthy, say NO to sugar!"

    if len(msgString) > 0:
      messageBox_label.value = msgString
      showPage("MessageBox")
  btnChoose.on_click(on_btnChoose_button_clicked)

  # display the objects
  vb1Col2_layout = widgets.Layout(display='flex-start', flex_flow='col', align_items='flex-start', border='1px solid grey')
  vb1Col2 = widgets.VBox([chooseTitle, rbGroup, btnChoose], layout=vb1Col2_layout)
  grid2[1:6,1] = vb1Col2


  ## Combo Box (static) / Reply
  replyTitle = Label("Feeling hungry?")
  cmb1 = widgets.Dropdown(options=["Yes! Very hungry!!!", "Just a bit...", "No, I'm full."], value=None)
  btnReply = widgets.Button(description="Reply")

  # define the event handlers for the buttons when they are being clicked
  def on_btnReply_button_clicked(b):
    messageBox_label.value = str(cmb1.value)
    showPage("MessageBox")
  btnReply.on_click(on_btnReply_button_clicked)

  # display the objects
  vb2Col1_layout = widgets.Layout(display='flex-start', flex_flow='col', align_items='flex-start', border='1px solid grey')
  vb2Col1 = widgets.VBox([replyTitle, cmb1, btnReply], layout=vb2Col1_layout)
  grid2[6:8,0] = vb2Col1


  ## Combo Box (dynamic) / Pick
  pickTitle = Label("Pick a student...")
  cmb2_options = []
  c.execute("SELECT student_name FROM PYSTUDENT")
  for row in c:
    cmb2_options.append(row[0])

  cmb2 = widgets.Dropdown(options=cmb2_options, value=None)
  btnPick = widgets.Button(description="Pick")

  # define the event handlers for the buttons when they are being clicked
  def on_btnPick_button_clicked(b):
    messageBox_label.value = "The winner of today is " + str(cmb2.value) + "!"
    showPage("MessageBox")
  btnPick.on_click(on_btnPick_button_clicked)

  # display the objects
  vb2Col2_layout = widgets.Layout(display='flex-start', flex_flow='col', align_items='flex-start', border='1px solid grey')
  vb2Col2 = widgets.VBox([pickTitle, cmb2, btnPick], layout=vb2Col2_layout)
  grid2[6:8,1] = vb2Col2

  display(grid2)

################ Sandbox Tab - End ################



########################################################
################ MessageBox Tab - Begin ################
########################################################

with tb_all.output_to('MessageBox'):

# create 1 label
  messageBox_label = Label("")

# create 1 button
  btnMsgBoxOK = widgets.Button(description="OK")

# define the event handlers for the two buttons when they are being clicked
  def on_btnMsgBoxOK_button_clicked(b):
    showPage("Sandbox")
  btnMsgBoxOK.on_click(on_btnMsgBoxOK_button_clicked)

# display the objects
  intro_row1HBox = widgets.HBox([messageBox_label])
  intro_row2HBox = widgets.HBox([btnMsgBoxOK])

  display(widgets.VBox([intro_row1HBox, intro_row2HBox]))

################ MessageBox Tab - End ################


showPage("Introduction")
```
